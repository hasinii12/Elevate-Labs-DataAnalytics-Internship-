```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Netflix.csv', encoding='latin-1')
df
```

```python
df.isnull()
```

```python
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
# Example: Standardize 'Date' column to dd-mm-yyyy
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.strftime('%d-%m-%Y')
df

```

```python
# Rename all column headers to lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```
